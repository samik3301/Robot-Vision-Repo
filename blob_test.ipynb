```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#3.1. Consider an image with circular object/squared/rectangular objects with two different size
image = cv2.imread("/Users/samik/Desktop/Programming/Robot Vision Lab/blob_input.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(image, cmap='gray')
```

```python
#3.2. Extract any related 5 blob features 
#by area 
params = cv2.SimpleBlobDetector_Params()
params.filterByArea = True
params.minArea = 100
detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(image)
im_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (255,0,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(im_with_keypoints, cmap='gray')

#plot the keypoints
total_number_blobs_detected = len(keypoints)
print("Number of keypoints Detected: ", total_number_blobs_detected)
for i in range(len(keypoints)):
    print("Keypoint ", i+1, ": ", keypoints[i].pt, keypoints[i].size)
print("Size of the keypoints: ", keypoints[0].size)
```

```python
params = cv2.SimpleBlobDetector_Params()
#by threshold
params.minThreshold = 0
params.maxThreshold = 255
detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(image)
im_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (255,0,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(im_with_keypoints, cmap='gray')

#plot the keypoints
print("Number of keypoints Detected: ", len(keypoints))
for i in range(len(keypoints)):
    print("Keypoint ", i+1, ": ", keypoints[i].pt, keypoints[i].size)
print("Size of the keypoints: ", keypoints[0].size)
```

```python
params = cv2.SimpleBlobDetector_Params()

# Filter by Convexity
params.filterByConvexity = True
params.minConvexity = 0.1
params.maxConvexity = 1

detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(image)
im_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (255,0,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(im_with_keypoints, cmap='gray')

#plot the keypoints
convex_blobs = len(keypoints)
print("Number of keypoints Detected: ",convex_blobs)
for i in range(len(keypoints)):
    print("Keypoint ", i+1, ": ", keypoints[i].pt, keypoints[i].size)
print("Size of the keypoints: ", keypoints[0].size)
```

```python
number_of_circular_blobs = circular_blobs
print("The total number of circular blobs detected are:  ",number_of_circular_blobs)

number_of_rectangle_blobs = total_number_blobs_detected-convex_blobs
print("The total number of rectangular or square blobs detected are: ",number_of_rectangle_blobs)

number_of_convex_blobs = convex_blobs
print("The number of blobs that aren't circular or rectangular in nature: ",number_of_convex_blobs)
```
