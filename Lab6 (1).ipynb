```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg')
img_grayscaled = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

pixel_vals = img_grayscaled.reshape((-1,1))   #grayscale only has 1 parameter

pixel_vals = np.float32(pixel_vals)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 150, 0.95)

rows =2
columns =2
fig = plt.figure(figsize =(17,13))
fig.add_subplot(rows,columns,1)
retval, labels, centers = cv2.kmeans(pixel_vals, 2, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image_k_2 = segmented_data.reshape((img_grayscaled.shape))
plt.imshow(segmented_image_k_2)
plt.axis('off')
plt.title("Segmented image for k=2")
fig.tight_layout()
fig.add_subplot(rows,columns,2)
retval, labels, centers = cv2.kmeans(pixel_vals, 3, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image_k_3 = segmented_data.reshape((img_grayscaled.shape))
plt.imshow(segmented_image_k_2)
plt.axis('off')
plt.title("Segmented image for k=3")
fig.tight_layout()
fig.add_subplot(rows,columns,3)
retval, labels, centers = cv2.kmeans(pixel_vals, 5, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image_k_5 = segmented_data.reshape((img_grayscaled.shape))
plt.imshow(segmented_image_k_5)
plt.axis('off')
plt.title("Segmented image for k=5")
fig.tight_layout()
fig.add_subplot(rows,columns,4)
retval, labels, centers = cv2.kmeans(pixel_vals, 7, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image_k_7 = segmented_data.reshape((img_grayscaled.shape))
plt.imshow(segmented_image_k_7)
plt.axis('off')
plt.title("Segmented image for k=7")
fig.tight_layout()
```
