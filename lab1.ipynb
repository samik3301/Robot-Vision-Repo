finding coordinates of contour

```python
# Python code to find the co-ordinates of the contours detected in an image.
import numpy as np
import cv2
font = cv2.FONT_HERSHEY_COMPLEX
img2 = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220611-151240_Comic_Screen.jpg', cv2.IMREAD_COLOR)
img = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220611-151240_Comic_Screen.jpg', cv2.IMREAD_GRAYSCALE)
_, threshold = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY)
contours, _= cv2.findContours(threshold, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours :
	approx = cv2.approxPolyDP(cnt, 0.009 * cv2.arcLength(cnt, True), True)
	cv2.drawContours(img2, [approx], 0, (0, 0, 255), 5)
	n = approx.ravel()
	i = 0
	for j in n :
		if(i % 2 == 0):
			x = n[i]
			y = n[i + 1]
			string = str(x) + " " + str(y)
			if(i == 0):
				cv2.putText(img2, "Arrow tip", (x, y),font, 0.5, (255, 0, 0))
			else:
				cv2.putText(img2, string, (x, y),font, 0.5, (0, 255, 0))
		i = i + 1
cv2.imshow('image2', img2)
if cv2.waitKey(0) & 0xFF == ord('q'):
	cv2.destroyAllWindows()
```

Piecewise Linear transformation functions

```python
import cv2
import numpy as np

# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, s1, r2, s2):
	if (0 <= pix and pix <= r1):
		return (s1 / r1)*pix
	elif (r1 < pix and pix <= r2):
		return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
	else:
		return ((255 - s2)/(255 - r2)) * (pix - r2) + s2
img = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg')
r1 = 70
s1 = 0
r2 = 140
s2 = 255
pixelVal_vec = np.vectorize(pixelVal)
contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)
cv2.imwrite('contrast_stretch.jpg', contrast_stretched)
```

Image Registeration

```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
img1_color = cv2.imread("sahil1.jpg") # Image to be aligned.
img2_color = cv2.imread("sahil2.jpg") # Reference image.
img1 = cv2.cvtColor(img1_color, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2_color, cv2.COLOR_BGR2GRAY)
height, width = img2.shape
orb_detector = cv2.ORB_create(5000)
kp1, d1 = orb_detector.detectAndCompute(img1, None)
kp2, d2 = orb_detector.detectAndCompute(img2, None)
matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = matcher.match(d1, d2)
matches = matches[:int(len(matches)*0.9)]
no_of_matches = len(matches)
p1 = np.zeros((no_of_matches, 2))
p2 = np.zeros((no_of_matches, 2))

for i in range(len(matches)):
    p1[i, :] = kp1[matches[i].queryIdx].pt
    p2[i, :] = kp2[matches[i].trainIdx].pt
homography, mask = cv2.findHomography(p1, p2, cv2.RANSAC)
transformed_img = cv2.warpPerspective(img1_color, homography, (width, height))
cv2.imwrite('gigachad.jpg', transformed_img)
```
