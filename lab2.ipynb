```python
# Python program to illustrate arithmetic operation of addition of two images
import cv2
import numpy as np
image1 = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220624-040151_Tachiyomi.jpg')
image2 = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220624-063257_Tachiyomi.jpg')
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
cv2.imshow('Weighted Image', weightedSum)

print("Image1 size before addition: ",image1.shape)
print("Image2 size before addition: ",image2.shape)
print("Added image size : ",weightedSum.shape)
print("Memory of the image1 before addition: ",image1.size)
print("Memory of the image2 before addition: ",image2.size)
print("Memory of the image after addition: ",weightedSum.size)

#231x500 make sure the dimensions of the 2 images are same for the code snippet to work
# De-allocate any associated memory usage
if cv2.waitKey(0) & 0xff == 27:
	cv2.destroyAllWindows()
```

```python
# Python program to illustrate arithmetic operation of subtraction of two images
import cv2
import numpy as np
image1 = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220624-040151_Tachiyomi.jpg')
image2 = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220624-063257_Tachiyomi.jpg')
weightedSub = cv2.subtract(image1, image2)
cv2.imshow('Subtracted Image', weightedSub)

print("Image1 size before subtraction: ",image1.shape)
print("Image2 size before subtraction: ",image2.shape)
print("Subtracted image size : ",weightedSub.shape)
print("Memory of the image1 before subtraction: ",image1.size)
print("Memory of the image2 before subtraction: ",image2.size)
print("Memory of the image after subtraction: ",weightedSub.size)

#231x500 make sure the dimensions of the 2 images are same for the code snippet to work
# De-allocate any associated memory usage
if cv2.waitKey(0) & 0xff == 27:
	cv2.destroyAllWindows()
    
```

```python
# Blurring image in OpenCV using python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\Screenshot_20220611-151240_Comic_Screen.jpg')
# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (7, 7), 0)
cv2.imwrite('Gaussian_Output_image.jpg',Gaussian)
print("Image size before Gaussian Blur: ",image.shape)
print("Image size after Gaussian Blur: ",Gaussian.shape)
print("Memory of the image before Gaussian Blur: ",image.size)
print("Memory of the image after Gaussian Blur: ",Gaussian.size)
print("\n")
# Median Blur
median = cv2.medianBlur(image, 5)
cv2.imwrite('Median_output_image.jpg',median)
print("Image size before Median Blur: ",image.shape)
print("Image size after Median Blur: ",median.shape)
print("Memory of the image before Median Blur: ",image.size)
print("Memory of the image after Median Blur: ",median.size)
print("\n")
# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imwrite('Bilateral_Output_image.jpg',bilateral)
print("Image size before Bilateral Blur: ",image.shape)
print("Image size after Bilateral Blur: ",bilateral.shape)
print("Memory of the image before Bilateral Blur: ",image.size)
print("Memory of the image after Bilateral Blur: ",bilateral.size)
print("\n")
rows =2
columns =2
fig = plt.figure(figsize =(6,11))
fig.add_subplot(rows,columns,1)
plt.imshow(image)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(Gaussian)
plt.axis('off')
plt.title("Gaussian Blurring")
fig.add_subplot(rows,columns,3)
plt.imshow(median)
plt.axis('off')
plt.title("Median Blurring")
fig.add_subplot(rows,columns,4)
plt.imshow(bilateral)
plt.axis('off')
plt.title("Bilateral Blurring")
```

```python
#Logarithmic transform
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg')
c = 255/(np.log(1 + np.max(img)))
log_transformed = c * np.log(1 + img)
log_transformed = np.array(log_transformed, dtype = np.uint8)
print("Image size before enhancement: ",img.shape)
print("Image size after enhancement: ",log_transformed.shape)
print("Memory of the image before enhancement: ",img.size)
print("Memory of the image after enhancemnet: ",log_transformed.size)
cv2.imwrite('log_transformed.jpg',log_transformed)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(log_transformed)
plt.axis('off')
plt.title("Enhanced Image after Logarithmic transform")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg")
gamma = 2.2
gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8')
cv2.imwrite('gamma_corrected_2.2.jpg',gamma_corrected)
print("Image size before enhancement: ",img.shape)
print("Image size after enhancement: ",gamma_corrected.shape)
print("Memory of the image before enhancement: ",img.size)
print("Memory of the image after enhancemnet: ",gamma_corrected.size)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(gamma_corrected)
plt.axis('off')
plt.title("Enhanced Image after Gamma correction")
```
