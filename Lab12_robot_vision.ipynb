```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/samik/Desktop/Programming/Robot Vision Lab/cloud.jpeg")
grayscale = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
(thresh, thresholded_img) = cv2.threshold(grayscale, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded_img,cmap="gray")
plt.title("Thresholded image - Binary image")
cv2.imwrite("O.jpg",thresholded_img)
```

```python
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(thresholded_img,kernel,iterations = 1)
plt.imshow(erosion,cmap="gray")
plt.title("Eroded image")
cv2.imwrite("O2.jpg",erosion)
```

```python
dilation = cv2.dilate(thresholded_img,kernel,iterations = 1)
plt.imshow(dilation,cmap="gray")
plt.title("Dilated image")
cv2.imwrite("O1.jpg",dilation)
```

```python
img2 = cv2.imread("/Users/samik/Desktop/Programming/Robot Vision Lab/Screenshot 2022-11-04 at 12.15.24 PM.png")
grayscale2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
(thresh, thresholded_img2) = cv2.threshold(grayscale2, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded_img2,cmap="gray")
plt.title("Thresholded Image having words (considered as separated by thin lines)")
cv2.imwrite("O3.jpg",thresholded_img2)
```

```python
opening = cv2.morphologyEx(thresholded_img2, cv2.MORPH_OPEN, kernel)
plt.imshow(opening,cmap="gray")
plt.title("Opening morphological operation on image")
cv2.imwrite("O4.jpg",opening)
```

```python
closing = cv2.morphologyEx(thresholded_img2, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing,cmap="gray")
plt.title("Closing morphological operation on image")
cv2.imwrite("O5.jpg",closing)
```
