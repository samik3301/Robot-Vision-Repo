Image Enhancement by adding a constant value of '90' to the first half of the image.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\bckg.png")
gray_scale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_scale[:gray_scale.shape[0]//2]+=90
cv2.imwrite('Added90.jpg',gray_scale)
print("Image size before enhancement: ",img.shape)
print("Image size after enhancement: ",gray_scale.shape)
print("Memory of the image before enhancement: ",img.size)
print("Memory of the image after enhancemnet: ",gray_scale.size)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(gray_scale)
plt.axis('off')
plt.title("Enhanced Image after addition")
```

Image enhancement using Gamma Correction

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\bckg.png")
gamma = 0.4
gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8')
print("Image size before enhancement: ",img.shape)
print("Image size after enhancement: ",gamma_corrected.shape)
print("Memory of the image before enhancement: ",img.size)
print("Memory of the image after enhancemnet: ",gamma_corrected.size)
cv2.imwrite('Gamma_corrected_0.4.jpg',gamma_corrected)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(gamma_corrected)
plt.axis('off')
plt.title("Enhanced Image after Gamma correction")
```

Dynamic Range Improvement Enhancement on Image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\bckg.png")
gray_scale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dynamic_range = 100
dynamic_range_corrected = gray_scale * (255.00/ dynamic_range)
cv2.imwrite('Dynamic_range_improve.jpg',dynamic_range_corrected)
print("Image size before enhancement: ",img.shape)
print("Image size after enhancement: ",dynamic_range_corrected.shape)
print("Memory of the image before enhancement: ",img.size)
print("Memory of the image after enhancemnet: ",dynamic_range_corrected.size)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(dynamic_range_corrected)
plt.axis('off')
plt.title("Enhanced Image after Dynamic Range Correction")
```
