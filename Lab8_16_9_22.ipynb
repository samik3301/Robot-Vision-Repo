```python
#Roberts edge detection method
import cv2 
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
  
roberts_cross_v = np.array( [[1, 0 ],
                             [0,-1 ]] )
  
roberts_cross_h = np.array( [[ 0, 1 ],
                             [ -1, 0 ]] )
  
img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg",0).astype('float64')
img/=255.0
vertical = ndimage.convolve( img, roberts_cross_v )
horizontal = ndimage.convolve( img, roberts_cross_h )
  
edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img*=255
cv2.imwrite("roberts_edge.jpg",edged_img)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
fig.add_subplot(rows,columns,2)
plt.imshow(edged_img)
plt.axis('off')
plt.title("Robert's Edge detected Image")
```

```python
#sobel edge detection method (using inbuilt functions)
import cv2 
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt

img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg",0).astype('float64') #grayscaled image
# Sobel Edge Detection using inbuilt functions
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
cv2.imwrite("Sobels_edge_usingInbuilt.jpg",sobelxy)

rows =2
columns =2
fig = plt.figure(figsize =(17,13))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image(Grayscaled)")
fig.tight_layout()

fig.add_subplot(rows,columns,2)
plt.imshow(sobelx)
plt.axis('off')
plt.title("Sobel edge detection on X axis")
fig.tight_layout()

fig.add_subplot(rows,columns,3)
plt.imshow(sobely)
plt.axis('off')
plt.title("Sobel edge detection on Y axis")
fig.tight_layout()

fig.add_subplot(rows,columns,4)
plt.imshow(sobelxy)
plt.axis('off')
plt.title("Sobel edge detection on X and Y axis")
fig.tight_layout()


```

```python
#prewitt's edge detection method
import cv2 
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
prewitts_cross_v = np.array( [[-1, 0, 1 ],
                             [-1,0,1 ],
                             [-1,0,1]] )
  
prewitts_cross_h = np.array( [[ 1,1,1 ],
                             [ 0,0, 0 ],
                             [-1,-1,-1]] )


img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg",0).astype('float64') #grayscaled image

img/=255.0
vertical = ndimage.convolve( img, prewitts_cross_v )
horizontal = ndimage.convolve( img, prewitts_cross_h )
  
edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img*=255
cv2.imwrite("Prewitts_edge.jpg",edged_img)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image(Grayscaled)")
fig.add_subplot(rows,columns,2)
plt.imshow(edged_img)
plt.axis('off')
plt.title("Prewitts's Edge detected Image")

```

```python
#Sobels edge detection [Hardcoded with kernels(horizontal and vertical)]
import cv2 
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
sobel_cross_v = np.array( [[-1, 0, 1 ],
                             [-2,0,2 ],
                             [-1,0,1]] )
  
sobel_cross_h = np.array( [[ 1,2,1 ],
                             [ 0,0, 0 ],
                             [-1,-2,-1]] )


img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg",0).astype('float64') #grayscaled image

img/=255.0
vertical = ndimage.convolve( img, sobel_cross_v )
horizontal = ndimage.convolve( img, sobel_cross_h )
  
edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img*=255
cv2.imwrite("Sobel_edge.jpg",edged_img)
rows =1
columns =2
fig = plt.figure(figsize =(20,15))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image(Grayscaled)")
fig.add_subplot(rows,columns,2)
plt.imshow(edged_img)
plt.axis('off')
plt.title("Sobel's Edge detected Image")
```

```python
#Edge subtraction (for all-Roberts, Sobel,Pritts)
import cv2 
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
img = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\eclipse_HD.jpg",0).astype('float64') #grayscaled image
Roberts_edge = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\Lab8\\roberts_edge.jpg",0).astype('float64')
Sobel_edge = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\Lab8\\Sobel_edge.jpg",0).astype('float64') #have used hardcoded Sobel instead of inbuilt functions
Prewitts_edge = cv2.imread("E:\\VSC_workspace\\robot_vision_lab\\Lab8\\Prewitts_edge.jpg",0).astype('float64')

Subtracted_edge_Roberts = cv2.subtract(img,Roberts_edge)
Subtracted_edge_Sobel = cv2.subtract(img,Sobel_edge)
Subtracted_edge_Prewitts = cv2.subtract(img,Prewitts_edge)

rows =2
columns =2
fig = plt.figure(figsize =(17,13))
fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image(Grayscaled)")
fig.tight_layout()

fig.add_subplot(rows,columns,2)
plt.imshow(Subtracted_edge_Roberts)
plt.axis('off')
plt.title("Subtracted edges using Robert's edge detection")
fig.tight_layout()

fig.add_subplot(rows,columns,3)
plt.imshow(Subtracted_edge_Sobel)
plt.axis('off')
plt.title("Subtracted edges using Sobel's edge detection")
fig.tight_layout()

fig.add_subplot(rows,columns,4)
plt.imshow(Subtracted_edge_Prewitts)
plt.axis('off')
plt.title("Subtracted edges using Prewitt's edge detection")
fig.tight_layout()
```
